```
#importing the required libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```
#Reading the excel sheet using pandas
air_data = pd.read_excel('AirQualityUCI.xlsx')
```

```
#Here we are going to plot the graph using seasborn
import seaborn as sns
from sklearn.preprocessing import StandardScaler
#Here scaling is performed by using StandardScalar
scalar = StandardScaler()
sns.set(style='whitegrid', context='notebook')
#The plotting is done on features considered below
features_plot = ['C6H6(GT)', 'RH', 'AH', 'PT08.S1(CO)']

data_to_plot = air_data[features_plot]
data_to_plot = scalar.fit_transform(data_to_plot)
data_to_plot = pd.DataFrame(data_to_plot)

```

```
#Here we are plotting a pairplot
sns.pairplot(data_to_plot, size=2.0);
plt.tight_layout()
plt.show()
```

```
air_data.dropna(axis=0, how='all')
```

```
features = air_data
features
```

```
#Here we are dropping some coloumns which does not contains CO
features = features.drop('Date', axis=1)
features = features.drop('Time', axis=1)
features = features.drop('C6H6(GT)', axis=1)
features = features.drop('PT08.S4(NO2)', axis=1)
features
```

```
labels = air_data['C6H6(GT)'].values
labels
```

```
features = features.values
features
```

```
#!pip install scikit-learn==0.19.0
```

```
#Here we are taking train_test_split class from sklearn.cross_validation package
from sklearn.cross_validation import train_test_split
```

```
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.3)
```

```
#Here we are going to convert the numpyarray to csv file using below statement
np.savetxt("foo.csv", X_test, delimiter=",")
```

```
#Here we are using LinearRegression Model from sklearn.linear_model package
from sklearn.linear_model import LinearRegression
```

```
#Here we are creating a object for the LinearRegression Class
regressor = LinearRegression()
#Here we are training the model by using fit method
regressor.fit(X_train, y_train)
```

```
#here the predicted values are stored in p
p=regressor.predict(X_test)
```

```
#Here we are predicting the values of CO based on Other gases.
print("Predicted values:", p)
```

```
#Here we are comparing the predicted data and the actual data
print("R^2 score for liner regression: ", regressor.score(X_test,y_test))
```
